```python
class FPTree:
    class Node:
        def __init__(self, item, frequency, parent):
            self.item = item
            self.freq = frequency
            self.nodeLink = None
            self.parent = parent
            self.children = {}

        def increment_counter(self, frequency):
            self.freq += frequency
        
        def display_tree_list(self):
            print(f'item: {self.item}, freq: {self.freq}',end='')
            if len(self.children)>0:
              print(",[",end='')
            for c in self.children.values():
              print("[",end='')
              c.display_tree_list()
              if len(c.children)==0:
                print("]",end='')
            print("]",end='')

    def get_data(self, wikipages):
        self.data = []
        for page in wikipages:
            p_wiki = wiki_wiki.page(page)
            self.data.append(p_wiki.text.split(sep=' '))
        self.Dict = {}
        for i in self.data:
            self.Dict[frozenset(i)] = 1
        return self.Dict

    def createTree(self, dataset, freqThreshold=2):
        HeaderTable = {}
        for transaction in dataset:
            for item in transaction:
                HeaderTable[item] = HeaderTable.get(item, 0) + dataset[transaction]
        for k in list(HeaderTable):
            if HeaderTable[k] < freqThreshold:
                del (HeaderTable[k])

        freqItemSet = set(HeaderTable.keys())

        if len(freqItemSet) == 0:
            return None, None

        for k in HeaderTable:
            HeaderTable[k] = [HeaderTable[k], None]

        FPTree = self.Node('root', 1, None)
        for itemset, freq in dataset.items():
            frequent_transaction = {}
            for item in itemset:
                if item in freqItemSet:
                    frequent_transaction[item] = HeaderTable[item][0]
            if len(frequent_transaction) > 0:
                ordered_itemset = [v[0] for v in sorted(frequent_transaction.items(), key=lambda p: p[1], reverse=True)]
                self.updateTree(ordered_itemset, FPTree, HeaderTable, freq)
        return FPTree, HeaderTable

    def updateTree(self, itemset, FPTree, HeaderTable, freq):
        if itemset[0] in FPTree.children:
            FPTree.children[itemset[0]].increment_counter(freq)
        else:
            FPTree.children[itemset[0]] = self.Node(itemset[0], freq, FPTree)

            if HeaderTable[itemset[0]][1] == None:
                HeaderTable[itemset[0]][1] = FPTree.children[itemset[0]]
            else:
                self.update_NodeLink(HeaderTable[itemset[0]][1], FPTree.children[itemset[0]])

        if len(itemset) > 1:
            self.updateTree(itemset[1::], FPTree.children[itemset[0]], HeaderTable, freq)

    def update_NodeLink(self, Test_Node, Target_Node):
        while (Test_Node.nodeLink != None):
            Test_Node = Test_Node.nodeLink

        Test_Node.nodeLink = Target_Node

    def FPTree_uptransveral(self, leaf_Node, prefixPath):
        if leaf_Node.parent != None:
            prefixPath.append(leaf_Node.item)
            self.FPTree_uptransveral(leaf_Node.parent, prefixPath)

    def find_prefix_path(self, basePat, TreeNode):
        conditionalPatternBase = {}

        while TreeNode != None:
            prefixPath = []
            self.FPTree_uptransveral(TreeNode, prefixPath)
            if len(prefixPath) > 1:
                conditionalPatternBase[frozenset(prefixPath[1:])] = TreeNode.freq
            TreeNode = TreeNode.nodeLink

        return conditionalPatternBase

    def Mine_Tree(self, FPTree, HeaderTable, minSupport, prefix, frequent_itemset):
        bigL = [v[0] for v in sorted(HeaderTable.items(), key=lambda p: p[1][0])]
        for basePat in bigL:
            new_frequentset = prefix.copy()
            new_frequentset.add(basePat)
            frequent_itemset.append(new_frequentset)
            Conditional_pattern_bases = self.find_prefix_path(basePat, HeaderTable[basePat][1])
            Conditional_FPTree, Conditional_header = self.createTree(Conditional_pattern_bases, minSupport)
            if Conditional_header != None:
                self.Mine_Tree(Conditional_FPTree, Conditional_header, minSupport, new_frequentset, frequent_itemset)
```
